```python
import os
import json
import numpy as np
import pandas as pd
from collections import defaultdict, Counter
#from neo4j import GraphDatabase
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
import seaborn as sns
```

```python
root_dir = './../data/complimentary_raw_data/raw_catalog_data/athena_pt_samples'
# List to hold all the file paths
file_paths = []

# Walk through the directory structure
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith('.csv') and not filename.startswith('.'):
            # Construct the full file path and add to the list
            file_paths.append(os.path.join(dirpath, filename))
print(f"Total CSV files found: {len(file_paths)}")
```

```python
# List to hold DataFrames
dataframes = []

# Iterate through each file path with a progress bar and read the CSV file into a DataFrame
for file_path in tqdm(file_paths[:20], desc="Reading CSV files"):
    try:
        # Extract the filename without the extension
        filename = os.path.basename(file_path)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
     
        # Append the DataFrame to the list
        dataframes.append(df)
        
    except pd.errors.ParserError as e:
        print(f"Error parsing {file_path}: {e}")

# Concatenate all the DataFrames into a single DataFrame if there are any valid DataFrames
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    # Output the shape of the combined DataFrame
    num_rows, num_cols = combined_df.shape
    print(f"Shape of the combined DataFrame: Rows = {num_rows}, Columns = {num_cols}")

    # Save the combined DataFrame to a CSV file
    combined_df.to_csv('combined_data.csv', index=False)
    print(f"Combined DataFrame written to 'combined_data.csv'")
    
    # Show the columns of the combined DataFrame
    print("Columns of the combined DataFrame:")
    print(combined_df.columns)
    
    # Show the first few rows of the combined DataFrame
    print("First few rows of the combined DataFrame:")
    print(combined_df.head())
else:
    print("No valid CSV files were processed.")
```

```python
nan_df = combined_df.groupby('category').apply(lambda x: x.isna().sum())
nan_df.head(1)
```

```python
nan_df["total_records"] = combined_df['category'].value_counts()
nan_df.head(1)
```

```python
def get_valid_columns(row):
    valid_columns = []
    total_records = row['total_records']
    for col in nan_df.columns[1:-1]:  # Exclude 'category' and 'category_total_records'
        if row[col] / total_records < 0.1:  # Check for more than 90% non-NaN
            valid_columns.append(col)
    return valid_columns
```

```python
nan_df['attributes'] = nan_df.apply(get_valid_columns, axis=1)
```

```python
# Initialize a dictionary to store the unique values for each attribute within each category
unique_values_dict = defaultdict(lambda: defaultdict(int))
```

```python
# Iterate over each category and attribute to populate the dictionary
for category in categories:
    category_attributes = exploded_df[exploded_df['category'] == category]['attributes'].tolist()
    category_df = cleaned_df[cleaned_df['category'] == category]
    
    for attribute in attributes:
        if attribute in category_attributes:
            unique_values_dict[category][attribute] = category_df[attribute].nunique()
        else:
            unique_values_dict[category][attribute] = 0

# Convert the dictionary to a DataFrame
unique_values_df = pd.DataFrame(unique_values_dict).T

# Fill NaN values with 0 (in case there are any missing entries)
unique_values_df = unique_values_df.fillna(0)

# Save the unique values DataFrame to a CSV file
unique_values_df.to_csv('./../data/unique_values_per_category.csv')
```

```python
# Display the unique values DataFrame
print("Unique values per attribute for each category:")
unique_values_df.head()
```

```python



# Filter out columns with all values 0 or 1
filtered_df = unique_values_df.loc[:, ~unique_values_df.isin([0, 1]).all()]

```

```python
# Number of categories and attributes
categories = filtered_df.index
n_categories = len(categories)

# Set up the figure and axes for a 4x5 grid of subplots
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(20, 16), constrained_layout=True)
axes = axes.flatten()  # Flatten the 2D array of axes to 1D

# Create a bar plot for each category
for i, category in enumerate(categories):
    ax = axes[i]
    sns.barplot(x=filtered_df.columns, y=filtered_df.loc[category], ax=ax)
    ax.set_title(f'Category: {category}')
    ax.set_xlabel('Attributes')
    ax.set_ylabel('Count of Unique Values')
    ax.tick_params(axis='x', rotation=90)  # Rotate x-axis labels for better readability

# Hide any remaining empty subplots if there are fewer categories than subplots
for j in range(n_categories, len(axes)):
    axes[j].axis('off')

# Show the plot
plt.show()
```
