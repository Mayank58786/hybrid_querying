```python
import random

def random_select_context(lst):
    if isinstance(lst, list) and len(lst) > 2:
        # Always include the first element
        selected = [lst[0]]
        # Randomly decide how many additional values to select from the remaining elements
        num_values = random.randint(0, len(lst) - 1)  # 0 to len(lst) - 1 (inclusive)
        if num_values > 0:
            # Select from the rest of the list (excluding the first element)
            selected += random.sample(lst[1:], num_values)  # lst[1:] excludes the first element
        return selected
    return lst 

def random_select_filter(lst):
    if isinstance(lst, list) and len(lst) > 2:
        # Randomly decide how many values to select (at least 1)
        num_values = random.randint(2, len(lst))  # Select at least 1 value
        return random.sample(lst, num_values)  # Sample without any restriction
    return lst 
```

```python
# Apply the random selection function to 'context' and 'filter' columns
df_1['context'] = df['context'].apply(random_select_context)
df_1['filter'] = df['filter'].apply(random_select_filter)
```
